```python
df= pd.read_csv(r"C:\Users\Student\Desktop\Python\LABELED SNAP PROTEIN WITH WASH 280.CSV", encoding= 'utf-16',sep='\t', names=["Time", "UV 280", "UV 493"])

df.head(50)
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


# Melt the DataFrame to long format if not already in that format
df_long = pd.melt(df, id_vars=['Time'], value_vars=['UV 280', 'UV 493'], 
                  var_name='Variable', value_name='Value')

# Use seaborn.relplot with 'hue' parameter
sns.relplot(data=df_long, x="Time", y="Value", hue="Variable", kind="line")

# Set plot labels and title
plt.xlabel("Time (Minutes)")
plt.ylabel("mAU")
plt.title("Plot of UV 280 and UV 493 over Time")

# Show the plot
plt.show()
```
