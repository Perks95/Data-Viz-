```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Basic Graph

```python
x =[0,1,2,3,4,5]
y = [0,2,4,6,8,10]

plt.figure(figsize=(6,4), dpi = 300)
plt.plot(x,y, label ='2x', color = "red", linewidth=2, marker= ".", linestyle="--")
plt.title(" Our first Graph", fontdict = {'fontname': 'Comic Sans MS', "fontsize": 20})
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

#line number 2
x2 = np.arange(0,4.5,0.5)
plt.plot(x2,x2**2, "r", label = "X^2")

plt.legend()
plt.savefig("mygraph.png",dpi=300)

plt.show()
```

Bar Chart

```python
labels = ['A', 'B', "C"]
Values = [1,4,2]

bars = plt.bar(labels, Values)
patterns = [ '/','O', "*"]
for bar in bars :
    bar.set_hatch(patterns.pop(0))
#bars[0].set_hatch("/")
#bars[1].set_hatch("O")
#bars[2].set_hatch("*")

plt.figure(figsize=(6,4))
plt.show()
```

```python
fifa=pd.read_csv(r"C:\Users\HP Demo\OneDrive - Stellenbosch University\Desktop\Python\Python\matplotlib_tutorial-master\fifa_data.csv")
```

```python
fifa.head()
```

```python
bins =[40,50,60,70,80,90,100]
plt.hist(fifa.Overall, bins= bins, color ="red" )
plt.title("Distribution of Player Skills in Fifa")
plt.xticks(bins)
plt.ylabel("Number of Players")
plt.xlabel("Overall rating")
plt.show()
```

```python
Left = fifa.loc[fifa["Preferred Foot"]=='Left'].count()[0]
right = fifa.loc[fifa["Preferred Foot"]=='Right'].count()[0]
print(Left)
right
labels = ["Left", 'Right']

plt.pie([Left,right], labels= labels, autopct='%.2f%%')
plt.title("Foot Preferrence in Fifa")
plt.show()
```

```python
fifa.Weight= [int(x.strip("lbs")) if type (x)==str else x for x in fifa.Weight]

plt.style.use("ggplot")

light = fifa.loc[fifa.Weight<125].count()[0]
light_medium= fifa.loc[(fifa.Weight>=125) & (fifa.Weight<150)].count()[0]
medium = fifa[(fifa.Weight>=150) & (fifa.Weight<200)].count()[0]
medium_heavy = fifa[(fifa.Weight>=175)&(fifa.Weight<200)].count()[0]
heavy= fifa[(fifa.Weight>= 200)].count()[0]


Weights = [light, light_medium, medium, medium_heavy, heavy]
labels = [ "Under 125", "125-150", '150-175','175-200', 'Over 200' ]
explode = (.1,.1,.1,.1,.1)

plt.pie(Weights, labels= labels, autopct="%.2f%%", pctdistance= 1.5, explode=explode)
plt.title("Weight Distribution of Fifa Players (in Lbs)")
plt.show()
```

```python
plt.style.use('default')
plt.figure(figsize=(5,8))

barcelona = fifa.loc[fifa.Club=="FC Barcelona"]["Overall"]
madrid = fifa.loc[fifa.Club=="Real Madrid"]["Overall"]
revs = fifa.loc[fifa.Club=="New England Revolution"]["Overall"]

labels = ['FC Barcelona', 'Real Madrid', 'New England Rev']

plt.boxplot([barcelona, madrid,revs], labels = labels)
#for box in boxes['boxes']:
    #box.set(color= "#4286f4", linewidth= 2)

plt.title('Professional Football Team comparisons')
plt.ylabel("Fifa Overall Rating")
plt.xlabel("Football Teams")

plt.show()
```
