```python
titan_drop= titanic.drop(['embarked', 'class', 'who', 'deck', 'embark_town','alone'],axis=1)
titan_drop.head()
```

```python
titan= titanic.select_dtypes(include=['int64', 'float64'])
titan.corr()
sns.heatmap(titan.corr(), annot=True)
plt.show()
```

```python
titan_drop.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Encode the binary categorical columns
titan_drop['sex'] = label_encoder.fit_transform(titan_drop['sex'])
titan_drop['adult_male'] = label_encoder.fit_transform(titan_drop['adult_male'])
titan_drop['alive'] = label_encoder.fit_transform(titan_drop['alive'])
```

```python
titan_drop.head()
```

```python
sns.heatmap(titan_drop.corr(), annot=True)
```
